# Validating the error of the mean with the Monte Carlo method

using simuatlions to test and assess a given problem. 

### $\sigma_{\overline{x}} = \frac{\sigma}{\sqrt{N}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import my_plot_style as mps
mps.my_plot_style()
```

```python
mu = 1
sigma = 1
n_simulation = 1000
mean_value = np.zeros(n_simulation)
n_point = 1000
for i_sim in range(0,n_simulation):
    random_data2 = np.random.normal(mu,sigma,n_point)
    mean_value[i_sim]=np.mean(random_data2)   
```

```python
plt.hist(mean_value)
plt.xlabel('mean x value')
print('Error of the mean (simulation)=',np.std(mean_value))
print('Error of the mean =',sigma/np.sqrt(n_point))
```

```python
''' I need to write this part in class'''

mu = 1
sigma = 1
n_simulation = 1000

n_point = np.arange(100,5000,100)
std_n = []
for i_point in range(0,len(n_point)):
    mean_value = np.zeros(n_simulation)
    for i_sim in range(0,n_simulation):
        random_data2 = np.random.normal(mu,sigma,n_point[i_point])
        mean_value[i_sim]=np.mean(random_data2)   
    
    std_n.append(np.std(mean_value))
```

```python
plt.figure(figsize=(8,8))
plt.plot(n_point,std_n,label='Monte Carlo method',ls='',marker='o')
plt.plot(n_point,sigma/np.sqrt(n_point),label='Theoretical expection')
plt.legend(frameon=False)
plt.yscale('log')
plt.xscale('log')
plt.ylabel('Error of the mean')
plt.xlabel('Number of data point')
```

# Cauchy distribution

```python

mu = 1
sigma = 1
n_simulation = 1000

n_point = np.arange(100,10000,100)
std_n = []
for i_point in range(0,len(n_point)):
    mean_value = np.zeros(n_simulation)
    for i_sim in range(0,n_simulation):
        random_data2 = np.random.standard_cauchy(n_point[i_point])
        mean_value[i_sim]=np.mean(random_data2)   
    
    std_n.append(np.std(mean_value))
```

```python
plt.figure(figsize=(8,8))
plt.plot(n_point,std_n,label='Cauchy distribution',ls='',marker='o')
plt.plot(n_point,3000*sigma/np.sqrt(n_point),label='Theoretical expection')
plt.legend(frameon=False)
plt.yscale('log')
plt.xscale('log')
plt.ylabel('Error of the mean')
plt.xlabel('Number of data point')
```

### Monte Carlo method checking 
### z = y-x
### $\sigma_{z}^{2} = \sigma_{x}^{2}+\sigma_{y}^{2}$

```python
mu_1 = 2
sigma_1 = 2
mu_2 = 5
sigma_2 = 7
variance_z = []
for i in range(0,1000):
    n_point = 100000
    x = np.random.normal(mu_1,sigma_1,n_point)
    y = np.random.normal(mu_2,sigma_2,n_point)
    z = y-x
    variance_z.append(np.var(z))
```

```python
plt.hist(variance_z)
plt.axvline(sigma_1**2+sigma_2**2,color='C1')
```
